# corelation between categorical columns
### chi-square test is a statistical test used to find the corelation between two categorical columns. so here we can use chi squae test.

```python
df57.head()
```

```python
df57.columns
```

```python
catogorical = [ 'job', 'marital', 'education','Gender', 'Insurance', 'housing', 'loan',
                            'contact','poutcome', 'Term Deposit']
df57_cat = df57[catogorical]
```

```python
df57_cat.head()
```

```python
import pandas as pd
from itertools import product
import numpy as np
import scipy.stats as ss
```

```python
## Let us split this list into two parts
cat_var1 = ( 'job', 'marital', 'education','Gender', 'Insurance', 'housing', 'loan',
                            'contact','poutcome', 'Term Deposit')
cat_var2 = ( 'job', 'marital', 'education','Gender', 'Insurance', 'housing', 'loan',
                            'contact','poutcome', 'Term Deposit')
```

```python
## Let us jump to Chi-Square test
## Creating all possible combinations between the above two variables list
cat_var_prod = list(product(cat_var1,cat_var2, repeat = 1))
cat_var_prod
```

```python
## Creating an empty variable and picking only the p value from the output of Chi-Square test
result = []
for i in cat_var_prod:
    if i[0] != i[1]:
        result.append((i[0],i[1],list(ss.chi2_contingency(pd.crosstab(
                            df57_cat[i[0]], df57_cat[i[1]])))[1]))
result
```

```python
chi_test_output = pd.DataFrame(result, columns = ['var1', 'var2', 'coeff'])

## Using pivot function to convert the above DataFrame into a crosstab
final_chi_output =chi_test_output.pivot(index='var1', columns='var2', values='coeff')
final_chi_output
```

```python
def highlight_low_values(value):
    if value <= 0.05 :
        return 'background-color: yellow'
    else:
        return ''

# Apply the styling function to the DataFrame
styled_df = final_chi_output.style.applymap(highlight_low_values)

# Display the styled DataFrame
styled_df
```

## There exists a relationship between two variables if p value ≤ 0.05. 
##   So from the above statistics, we can see that except gender variable all other catogorical variables has p_value <= 0.05  with the target 'Term deposit'.

 # 2. Predictive Model Building:

```python
# Split Dataset into Training set and Test set
```

```python
df59 = df58.copy()
```

```python
df59.drop(['Customer_number','AgeGroup'],axis=1,inplace=True)
```

```python
df59.head()
```

```python
# splitting the train and test data
X = df59.drop(['Term Deposit_new'],axis=1)
y = df59['Term Deposit_new']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

## Logistic regression

```python
# LOGISTIC REGRESSION:
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
# define the model
model = LogisticRegression()
# fit on the training set
model.fit(X_train, y_train)
# predict on test set
train_pred= model.predict(X_train)
ypred = model.predict(X_test)
# evaluate predictions
accuracy_train = accuracy_score(y_train,train_pred)
accuracy = accuracy_score(y_test, ypred)
log_cross_val_score =cross_val_score(estimator=LogisticRegression(),X=X_train, y=y_train, cv=5)
print('cross validation score for logistic regression:', log_cross_val_score)
print('train_Accuracy_score_Logistic_reg: %.2f' % (accuracy_train*100))
print('Test_Accuracy_score_Logistic_reg: %.2f' % (accuracy*100))
```

## Random forest classifier

```python
# RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=5, random_state=0)
clf.fit(X_train, y_train)
# predict on test set
train_pred= clf.predict(X_train)
ypred = clf.predict(X_test)
# evaluate predictions
accuracy_train = accuracy_score(y_train,train_pred)
accuracy = accuracy_score(y_test, ypred)
rf_cross_val_score =cross_val_score(estimator=RandomForestClassifier(),X=X_train, y=y_train, cv=5)
print('cross validation score for RandomForest:', rf_cross_val_score)
print('train_Accuracy_score_Random_forest: %.2f' % (accuracy_train*100))
print('Test_Accuracy_score_Random_Forest: %.2f' % (accuracy*100))
```

## Gradient boosting classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb.fit(X_train,y_train)
# predict on test set
train_pred= gb.predict(X_train)
ypred = gb.predict(X_test)
# evaluate predictions
accuracy_train = accuracy_score(y_train,train_pred)
accuracy = accuracy_score(y_test, ypred)
gb_cross_val_score =cross_val_score(estimator=GradientBoostingClassifier(),X=X_train, y=y_train, cv=5)
print('cross validation score for Gradient_Boosting:', gb_cross_val_score)
print('train_Accuracy_score_Random_forest: %.2f' % (accuracy_train*100))
print('Test_Accuracy_score_Random_Forest: %.2f' % (accuracy*100))
```
