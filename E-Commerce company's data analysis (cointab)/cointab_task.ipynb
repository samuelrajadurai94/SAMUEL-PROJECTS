```python
def fwd_and_rto_charge_cal(df) :
    if df['no_of_weightslabs_asperX']==1:
        if 'Forward charges' in df['Type of Shipment']:
            return df['Forward Fixed Charge']
        else:
            return df['Forward Fixed Charge'] + df['RTO Fixed Charge']
    else:
        if 'Forward charges' in df['Type of Shipment']:
            return df['Forward Fixed Charge'] + ((df['no_of_weightslabs_asperX']-1) * df['Forward Additional Weight Slab Charge'])
        else:
            return df['Forward Fixed Charge'] + ((df['no_of_weightslabs_asperX']-1) * df['Forward Additional Weight Slab Charge'])+df['RTO Fixed Charge'] +((df['no_of_weightslabs_asperX']-1) * df['RTO Additional Weight Slab Charge']) 
        
            
```
